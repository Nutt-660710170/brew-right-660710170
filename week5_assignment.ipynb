## 1 แสดง ชื่อmodel และ ราคา ของโทรศัพท์ที่มี Battery Capacity >= 4500

```python
query = """
select
  Model,
  Price,
  Battery_Capacity_mAh
from 'Mobile phone price.csv'
where Battery_Capacity_mAh >= 4500;
"""
```

```python
Senario_1 = con.execute(query).fetchdf()
```

## 2 แสดงโทรศัพท์ Xiaomi ที่มีราคาตั้งแต่ 200-500

```python
query = """
select
  Brand,
  Model,
  Price
from 'Mobile phone price.csv'
where Price >= 200
and Price <= 500
and Brand = 'Xiaomi';
"""
```

```python
Senario_2 = con.execute(query).fetchdf()
```

##6 หาว่าโทรศัพท์แต่ละ Brand มีทั้งหมดกี่รุ่น

##7 หาขนาดหน้าจอ (ScreenSize_inches) ที่ใหญ่ที่สุด และรุ่นที่ใช้

```python
query = """
select
  Model,
  ScreenSize_inches,
  Price
from 'Mobile phone price.csv'
order by ScreenSize_inches desc
limit 1;
"""
```

##8 หาโทรศัพท์ที่มีค่า Battery_Capacity_mAh มากที่สุดในแต่ละ Brand

```python
query = """
select
  Brand,
  any_value(Model) as Model,
  max(cast(Battery_Capacity_mAh as double)) as Max_Battery
from 'Mobile phone price.csv'
group by Brand;
"""
```

```python
Senario_8 = con.execute(query).fetchdf()
```
